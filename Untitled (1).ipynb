```python
import numpy as np
import pandas as pd
from pathlib import Path
import os.path

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split

import tensorflow as tf
from sklearn.metrics import confusion_matrix,classification_report
```

```python
%pwd
```

```python
image_dir=Path(r"/Users/harshinisaidonepudi/Downloads/food_image/Indian_Food_Images/Indian Food Images")
```

```python
filepaths=list(image_dir.glob(r'**/*.jpg'))
labels=list(map(lambda x:os.path.split(os.path.split(x)[0])[1],filepaths))
```

```python
labels
```

```python
filepaths=pd.Series(filepaths,name='Filepath').astype(str)
labels=pd.Series(labels,name='Label')

images=pd.concat([filepaths,labels],axis=1)
```

```python
images
```

```python
train_df,test_df=train_test_split(images,train_size=0.7,shuffle=True,random_state=1)
```

```python
train_generator=tf.keras.preprocessing.image.ImageDataGenerator(
    preprocessing_function=tf.keras.applications.mobilenet_v2.preprocess_input,
    validation_split=0.2
)

test_generator=tf.keras.preprocessing.image.ImageDataGenerator(
    preprocessing_function=tf.keras.applications.mobilenet_v2.preprocess_input
)
```

```python
train_images=train_generator.flow_from_dataframe(
    dataframe=train_df,
    x_col='Filepath',
    y_col='Label',
    target_size=(224,224),
    color_mode='rgb',
    class_mode='categorical',
    batch_size=32,
    shuffe=True,
    seed=42,
    subset='training'
)

val_images=train_generator.flow_from_dataframe(
    dataframe=train_df,
    x_col='Filepath',
    y_col='Label',
    target_size=(224,224),
    color_mode='rgb',
    class_mode='categorical',
    batch_size=32,
    shuffe=True,
    seed=42,
    subset='training'
)

test_images=test_generator.flow_from_dataframe(
    dataframe=test_df,
    x_col='Filepath',
    y_col='Label',
    target_size=(224,224),
    color_mode='rgb',
    class_mode='categorical',
    batch_size=32,
    shuffe=False
)
```

```python
pretrained_model=tf.keras.applications.MobileNetV2(
    input_shape=(224,224,3),
    include_top=False,
    weights='imagenet',
    pooling='avg'
)
pretrained_model.trainable=False
```

```python
inputs=pretrained_model.input
x=tf.keras.layers.Dense(80,activation='relu')(pretrained_model.output)
x=tf.keras.layers.Dense(80,activation='relu')(x)

outputs=tf.keras.layers.Dense(80,activation='softmax')(x)

model=tf.keras.Model(inputs,outputs)

print(model.summary())
```

```python
model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
history=model.fit(
    train_images,
    validation_data=val_images,
    epochs=10
)
```

```python
results=model.evaluate(test_images,verbose=0)
print('Tset accuracy:{:.2f}%'.format(results[1]*100))
```

```python
prediction=np.argmax(model.predict(test_images),axis=1)

cn=confusion_matrix(test_images.labels,prediction)
clr=classification_report(test_images.labels,prediction,target_names=test_images.class_indices)
```

```python
print(clr)
```

```python
plt.figure(figsize=(30,30))
sns.heatmap(cn,annot=True,fmt='g',vmin=0,cmap='Blues',cbar=False)
plt.xticks(ticks=np.arange(78)+0.5,labels=test_images.class_indices,rotation=90)
plt.yticks(ticks=np.arange(78)+0.5,labels=test_images.class_indices,rotation=0)
plt.xlabel("predicted")
plt.ylabel("actual")
plt.title("confusion_matrix")
plt.show()
```
